### Loading Dependencies 

```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
```

### Reading Data from CSV Into Pandas

```python
# Name of the Projects_Data CSV file
projfile = "Resources/projects_data.csv"
```

```python
# Read the Projects file
projdf = pd.read_csv(projfile, low_memory=False)
```

```python
# Preview of the DataFrame
projdf.head()
```

```python
# Select only certain columns for analysis based on completeness of data
newprojdf = projdf[['id','regionname','countryname','prodline','lendinginstr','lendinginstrtype',\
                            'envassesmentcategorycode','projectstatusdisplay','status','project_name',\
                            'boardapprovaldate', 'closingdate','lendprojectcost','totalamt','grantamt','url']]
```

### Filtering and Cleansing Data in the Panda 

```python
# Filter and select rows where columns match selection criteria for LendingInstrType and EnvAssesmentCategoryCode
filterdf = newprojdf.loc[((newprojdf["lendinginstrtype"] == "AD") | (newprojdf["lendinginstrtype"] == "IN" )) \
                             & ((newprojdf["envassesmentcategorycode"] == 'A') \
                                  | (newprojdf["envassesmentcategorycode"] == 'B') \
                                  | (newprojdf["envassesmentcategorycode"] == 'C')  \
                                  |  (newprojdf["envassesmentcategorycode"] == 'D')) \
                                  & newprojdf["projectstatusdisplay"] != '' ]

# Drop rows where the EnvAssesmentCategoryCode has 'NaN' as value
filterdf = filterdf.dropna(subset=["envassesmentcategorycode"])

# Display new Dataframe
filterdf
```

```python
# TotalAmt, GrantAmt and LenderProjectCost are all String Data
# Convert them to numeric by first removing the ',' character by using the Split function

filterdf['totalamt'] = filterdf['totalamt'].str.replace(',','')
filterdf['grantamt'] = filterdf['grantamt'].str.replace(',','')
filterdf['lendprojectcost'] = filterdf['lendprojectcost'].str.replace(',','')

# Convert the columns to Float using the astype function
filterdf['totalamt'] = filterdf['totalamt'].astype(float)
filterdf['grantamt'] = filterdf['grantamt'].astype(float)
filterdf['lendprojectcost'] = filterdf['lendprojectcost'].astype(float)

# Display Dataframe to view the results
filterdf
```

```python
# Country names are repeated with ';' separating the names
# Using Split function split the names into multiple panda elements

countrydf  = pd.DataFrame(filterdf['countryname'].str.split(';', expand=True))

# Board Approval Date and Close Date strings where the Date is concatenated with Time with a 'T' separating them
# Using Split function split the Date and Time into multiple panda elements

bapdtdf = pd.DataFrame(filterdf['boardapprovaldate'].str.split('T', expand=True))
closdtdf = pd.DataFrame(filterdf['closingdate'].str.split('T', expand=True))
```

```python
# From the main dataframe, drop the erroenous columns - 'countryname','boardapprovaldate' and 'closingdate'
filterdf = filterdf.drop(columns=['countryname','boardapprovaldate','closingdate'])

# Replace these columns with the 1st element of each of the pandas created in the previous frame using the Split function
filterdf["countryname"] = countrydf[0]
filterdf["boardapprovaldate"] = bapdtdf[0]
filterdf["closingdate"] = closdtdf[0]
```

```python
# Remove additional columns such as Goal and ProjectStatusDisplay and copy data in a new DataFrame 
finaldf = filterdf[['id','project_name','regionname','countryname','prodline','lendinginstr','lendinginstrtype',\
                            'envassesmentcategorycode','status','boardapprovaldate', 'closingdate','lendprojectcost',\
                            'totalamt','grantamt','url']]

# Display Dataframe to view the results
finaldf
```

### Finalizing the Dataframe for Analysis 

```python
# Reorder and Rename the column headers to make the Dataframe more readable
finaldf = finaldf.rename({'id':'Project_ID','project_name':'Project_Name','regionname':'Region_Name',\
                          'countryname':'Country_Name', 'prodline':'Product_Line','lendinginstr':'Lending_Instrument',\
                          'lendinginstrtype':'Lending_Instrument_Type','envassesmentcategorycode':'Env_Assessment_Code',\
                          'status':'Project_Status','boardapprovaldate':'Board_Approval_Dt', \
                          'closingdate':'Closing_Date','lendprojectcost':'Lender_Cost',\
                          'totalamt':'Total_Amt','grantamt':'Grant_Amt','url':'Project_URL'}, axis =1)

# Display Dataframe to view the results
finaldf.head()
```

### Analysis of Closed Projects¶

```python
# Copy all the rows where the Project_Status is 'Closed' into a new Dataframe
ClosedProjdf = finaldf.loc[(finaldf["Project_Status"] == "Closed")]

# Display Dataframe to view the results
ClosedProjdf
```

```python
# Group the data by Country_Name
CloseByCountrydf = ClosedProjdf.groupby(["Country_Name"])

# Compute Average Total_Amount, Median Total_Amount and Sum of Total_Amount across all projects for a country
AvgTotalAmt = CloseByCountrydf["Total_Amt"].mean()
MedTotalAmt = CloseByCountrydf["Total_Amt"].median()
SumTotalAmt = CloseByCountrydf["Total_Amt"].sum()

# Create a new Dataframe with these values

CloseByCountry_Statdf = pd.DataFrame({"Average_Total_Amount": AvgTotalAmt,
                                    "Median_Total_Amount": MedTotalAmt,
                                     "Sum_Total_Amount": SumTotalAmt}) 

# Select only those rows where at least one of the measures - Average, Median or Sum is > 0
CloseByCountry_Finaldf = CloseByCountry_Statdf.loc[(CloseByCountry_Statdf['Average_Total_Amount'] != 0) \
                                                   |(CloseByCountry_Statdf['Median_Total_Amount'] != 0) \
                                                   |(CloseByCountry_Statdf['Sum_Total_Amount'] != 0)]
# Display Dataframe
CloseByCountry_Finaldf
```

```python
# Select the Top 5 countries in terms of the Sum of the Total Cost of the Project
CloseByCountry_Top5df = CloseByCountry_Finaldf.nlargest(n=5, columns=["Sum_Total_Amount"])

# Display the Results
CloseByCountry_Top5df
```

```python
# Create a Bar Graph for the Top 5 countries 

CloseByCountry_Top5df['Sum_Total_Amount'].plot(kind="bar")

plt.title("Top 5 Member Countries - Completed Projects")
plt.ylabel("Total Project Cost (in Billion USD)")

# Display the graph
plt.show()
```

```python
# Select the Bottom 5 countries in terms of the Sum of the Total Cost of the Project
CloseByCountry_Bottom5df = CloseByCountry_Finaldf.nsmallest(n=5, columns=["Sum_Total_Amount"])

#Display Results
CloseByCountry_Bottom5df
```

```python
# Create a Bar Graph for the Bottom 5 countries 

CloseByCountry_Bottom5df['Sum_Total_Amount'].plot(kind="bar")

plt.title("Lowest 5 Member Countries - Completed Projects")
plt.ylabel("Total Project Cost (in Billion USD)")

#Display Graph
plt.show()
```

### Analysis of Active Projects¶

```python
# Copy all the rows where the Project_Status is 'Active' into a new Dataframe
ActiveProjdf = finaldf.loc[(finaldf["Project_Status"] == "Active")]

# Display Dataframe to view the results
ActiveProjdf
```

```python
# Group the data by Region_Name
ActiveRegiondf = ActiveProjdf.groupby(["Region_Name"])

# Compute Average Total_Amount, Median Total_Amount and Sum of Total_Amount across all projects for a Region
AvgTotalAmt = ActiveRegiondf["Total_Amt"].mean()
MedTotalAmt = ActiveRegiondf["Total_Amt"].median()
SumTotalAmt = ActiveRegiondf["Total_Amt"].sum()

# Create a new Dataframe with these values
ActiveByRegion_Statdf = pd.DataFrame({"Average_Total_Amount": AvgTotalAmt,
                                    "Median_Total_Amount": MedTotalAmt,
                                     "Sum_Total_Amount": SumTotalAmt})

# Select only those rows where at least one of the measures - Average, Median or Sum is > 0
ActiveByRegion_Finaldf = ActiveByRegion_Statdf.loc[(ActiveByRegion_Statdf['Average_Total_Amount'] != 0) \
                                                   |(ActiveByRegion_Statdf['Median_Total_Amount'] != 0) \
                                                   |(ActiveByRegion_Statdf['Sum_Total_Amount'] != 0)]
# Display Dataframe
ActiveByRegion_Finaldf
```

```python
# Create a Bar Graph displaying Average Project Cost for each Region
ActiveByRegion_Finaldf['Average_Total_Amount'].plot(kind="bar")

plt.title("Average Cost By Region - Active Projects")
plt.ylabel("Average Project Cost (in Billion USD)")
plt.xlabel("Region")

# Display Graph
plt.show()
```
