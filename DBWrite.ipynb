### Load Dependencies 

```python
import pandas as pd
from sqlalchemy import create_engine
```

### Read Files Created By Upstream Programs

```python
csv_file = "Output/projects_data_output.csv"
projects_data_df = pd.read_csv(csv_file)
projects_data_df
```

```python
csv_file1 = r"Project2/Output/electricity_final.csv"
electricity_df = pd.read_csv(csv_file1)
electricity_df
```

```python
csv_file2 = r"Resources/population_data_clean.csv"
population_df = pd.read_csv(csv_file2)

population_final_df = population_df[['Country Name','Country Code','Indicator Name','Indicator Code','Year','Population']]
population_final_df
```

```python
csv_file3 = r"Project2/Output/gdp_final.csv"
gdp_final_df = pd.read_csv(csv_file3)
gdp_final_df
```

```python
csv_file4 = r"Resources/gdp_data_clean.csv"
gdp_df = pd.read_csv(csv_file4)

gdp_df = gdp_df[["Country Name","Country Code","Indicator Name","Indicator Code","Year","GDP"]]
gdp_df
```

```python
csv_file = "Output/active_projects_output.csv"
active_projects_df = pd.read_csv(csv_file)
active_projects_df
```

```python
csv_file = "Output/closed_projects_output.csv"
closed_projects_df = pd.read_csv(csv_file)
closed_projects_df
```

### Create Database Connection In PostgreSQL

```python
rds_connection_string = "postgres: @localhost:5432/WorldBank_ETL"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Write Dataframes into Database Tables 

```python
projects_data_df.to_sql(name='projects_data', con=engine, if_exists='append', index=False)
```

```python
electricity_df.to_sql(name='electricity', con=engine, if_exists='append', index=False)
```

```python
population_final_df.to_sql(name='population', con=engine, if_exists='append', index=False)
```

```python
gdp_final_df.to_sql(name='gdp_electricity', con=engine, if_exists='append', index=False)
```

```python
gdp_df.to_sql(name='gdp', con=engine, if_exists='append', index=False)
```

```python
active_projects_df.to_sql(name='active_projects', con=engine, if_exists='append', index=False)
```

```python
closed_projects_df.to_sql(name='closed_projects', con=engine, if_exists='append', index=False)
```
